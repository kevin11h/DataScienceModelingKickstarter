```python
def one_of_k_encode_parent_category(df):
    for parent_category in sorted(df.parent_category.unique()):
        colname = ''.join(parent_category.lower().split()).replace('amp;', '')
        df['its_' + colname] = map(lambda par_cat_val : 1 if par_cat_val == parent_category else 0,
                                   df.parent_category)
        df['not_' + colname] = 1 - df['its_' + colname]
        
def one_of_k_encode_child_category(df):
    for child_category in sorted(df.child_category.unique()):
        colname = ''.join(child_category.lower().split()).replace('amp;', '')
        df['its_' + colname] = map(lambda child_cat_val : 1 if child_cat_val == child_category else 0,
                                   df.child_category)
        df['not_' + colname] = 1 - df['its_' + colname]
        
one_of_k_encode_parent_category(DM_KS_DF)
one_of_k_encode_parent_category(JP_KS_DF)
one_of_k_encode_child_category(DM_KS_DF)
one_of_k_encode_child_category(JP_KS_DF)
```

### Factor Analysis

I want to study the relationship between the features, then formulate a model that captures that relationship.

Let's build and visualize a correlation matrix of the features! <br />
Each entry of the correlation matrix has the form:

### $$\begin{aligned}\rho_{X, Y} &= \mathrm{Corr}(X, Y) = {\mathrm{Cov}(X, Y) \over \sqrt{ Var(X) \mathrm{ } Var(Y) } } = {\mathrm{Cov}(X, Y) \over {\sigma_X \mathrm{ } \sigma_Y } } = {E[(X-\mu_X)(Y-\mu_Y)] \over \sigma_X\sigma_Y} \mathrm{,} \\ & \mathrm{where } -1 \leq \rho_{X, Y} \leq 1 \end{aligned}$$

```python
noncomputational_columns = ['name', 'parent_category', 'child_category']
is_computational = lambda c : c not in noncomputational_columns
doesnt_start_with_not = lambda c : not c.startswith('not')

def get_computational_columns(df):
    computational_columns = filter(lambda c : is_computational(c) and doesnt_start_with_not(c), df.columns)
    
    if 'successful' in computational_columns:
        i = computational_columns.index('successful')
        del computational_columns[i]
        computational_columns.append('successful')
        
    return computational_columns
        
def to_matrix_4_ML(df):
    computational_columns = get_computational_columns(df)
    return df[computational_columns].as_matrix()
```

```python
def center_feature_matrix(feature_matrix):
    mean_project = feature_matrix.mean(axis=0)
    return feature_matrix - mean_project

def get_normalization_factor(feature_matrix):
    n_samples, k_features = feature_matrix.shape
    norm = np.float64(n_samples-1)
    return norm

def calculate_correlation_matrix(feature_matrix):
    centered_feature_matrix = center_feature_matrix(feature_matrix)
    norm = get_normalization_factor(centered_feature_matrix)
    Cov = np.dot(centered_feature_matrix.T, centered_feature_matrix) / norm

    variances = np.diag(Cov)
    Corr = Cov / np.sqrt(np.outer(variances, variances).astype(np.float64))
    
    return Corr
```

```python
training_matrix = to_matrix_4_ML(TRAIN_DF)
Corr = calculate_correlation_matrix(training_matrix)
```

Numpy and Scipy have correlation, covariance, PCA and eigenvalue decomposition routines, and we might use them later on, but it is important to learn what they are calculating. Let's compare our function with numpy's corrcoef:

```python
Rho = np.corrcoef(training_matrix.T.astype(np.float64))
Rho.min(), Rho.max(), Corr.min(), Corr.max()
```

Another good check is to see if the eigenvalues of Corr/Rho are positive which means that it is indeed a positive semidefinite matrix

```python
def columns_2_labels_4_viz(columns):
    labels = []
    
    for c in columns:
        if str(c).startswith('its_'):
            c = c[len('its_'):]
        
        if str(c).endswith('_count'):
            c = c[0:-len('_count')]
            
        c = c[0].upper() + c[1:]
        labels.append(c)
        
    return labels    

trace = Heatmap(z=Corr,
                x=columns_2_labels_4_viz(get_computational_columns(TRAIN_DF)),
                y=columns_2_labels_4_viz(get_computational_columns(TRAIN_DF)),
                zauto=False,
                zmin = Corr.min(),
                zmax = Corr.max(),
                colorscale='YIOrRd',
                reversescale=True)

data = Data([trace])

title = 'Feature Correlations of KickStarter Projects in June 2012 with Categories'
data_src_url = 'http://www.appsblogger.com/behind-kickstarter-crowdfunding-stats/'
annotation_text = 'Source and for more information: <a href="%s">Jeanne on Appsblogger</a>' % data_src_url

layout = Layout(title=title,
                autosize=False,
                height=500,
                width=600,
                annotations=Annotations([Annotation(text=annotation_text,
                                         xref='paper',
                                         yref='paper',
                                         xanchor='right',
                                         yanchor='bottom',
                                         x=.75,
                                         y=1)])
                )

fig = Figure(data=data, layout=layout)

url = py.plot(fig, filename='ks_feature_corr_cat', auto_open=False)
print url
py.iplot(fig, filename='ks_feature_corr_cat')
```

```python
def exclude_categories(columns):
    return filter(lambda c : not str(c).startswith('its'), columns)

reduced_computational_columns = exclude_categories(get_computational_columns(TRAIN_DF))
reduced_training_matrix = TRAIN_DF[reduced_computational_columns].as_matrix()
Corr2 = calculate_correlation_matrix(reduced_training_matrix)
reduced_training_df = DataFrame(reduced_training_matrix, columns=reduced_computational_columns)
reduced_training_df.to_csv('reduced_training_matrix.csv', index=False)
reduced_training_df.head(len(reduced_computational_columns))
```

```python
def triangularize_symmetric_matrix(x, y, z):
    x_new = x[0:-1]
    y_new = y[-1:0:-1]
    tmp = np.empty(z.shape)
    tmp.fill(np.nan)
    for i in range(len(z)):
        tmp[i, 0 :i] = z[i, 0:i]
    tmp = np.flipud(tmp)
    z_new = tmp
    return (x_new, y_new, z_new)

(x, y, z) = triangularize_symmetric_matrix(columns_2_labels_4_viz(exclude_categories(get_computational_columns(TRAIN_DF))),
                                           columns_2_labels_4_viz(exclude_categories(get_computational_columns(TRAIN_DF))),
                                           Corr2)

trace = Heatmap(z=z,
                x=x,
                y=y,
                zauto=False,
                zmax = Corr2.max(),
                zmin = Corr2.min(),
                colorscale=[[0,'#CC0000'], [.5, '#CC00CC'], [1, '#3333CC']],
                reversescale=True,
                colorbar=ColorBar(titleside='bottom',
                                  title='   %f'%Corr2.min()))

data = Data([trace])

title = 'Feature Correlations of KickStarter Projects in June 2012'
data_src_url = 'http://www.appsblogger.com/behind-kickstarter-crowdfunding-stats/'
annotation_text = 'Source and for more information: <a href="%s">Jeanne on Appsblogger</a>' % data_src_url

layout = Layout(title=title,
                autosize=False,
                height=500,
                width=600,
                annotations=Annotations([Annotation(text=annotation_text,
                                         xref='paper',
                                         yref='paper',
                                         xanchor='right',
                                         yanchor='bottom',
                                         x=.75,
                                         y=1)])
                )
axis_style = {'autotick':False,'showgrid':False,'showline':False}
layout.update(xaxis=XAxis(axis_style, tickangle=90))
layout.update(yaxis=YAxis(axis_style))

fig = Figure(data=data, layout=layout)

url = py.plot(fig, filename='ks_feature_corr', auto_open=False)
print url
py.iplot(fig, filename='ks_feature_corr')
```
